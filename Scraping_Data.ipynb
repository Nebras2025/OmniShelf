# in this section we're going to Scrape Data from "Book to scrap" website

```python
# Importing Requests libary

import requests
```

```python
# Running a loop to retrieve data from 1-50 pages of the website

for i in range(1, 51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"htmls/page{i}.html", "w", encoding="utf-8") as f:
        f.write(a.text)
        print(f"Download page {i} successfully")
```

```python
# Importing beautifulsoup libary

from bs4 import BeautifulSoup
```

```python
# running a loop for reading the contents in the html file which(data) i scrape from the website

for i in range(1,51):
    with open(f"htmls/page{i}.html", "r", encoding="utf-8") as f:
        content = f.read()
```

```python
# creating a "Soup" for scraping data from the raw html

soup = BeautifulSoup(content, "html.parser")
```

```python
articles = soup.select("article.product_pod")
```

```python
# finding all the title, price & rating

import glob
from bs4 import BeautifulSoup

items = []
for file in sorted(glob.glob("htmls/*.html")):

    with open(file, encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    for book in soup.select("article.product_pod"):
        title = book.h3.a["title"]
        price = book.select_one(".price_color").text.split('£')[1]
        rating = book.select_one(".star-rating")["class"][1]

        items.append({
            "title" : title,
            "price" : price,
            "rating" : rating
        })      
```

```python
# creating a data frame for the 'items'

import pandas as pd

df = pd.DataFrame(items)
df.columns = ["Book name", "Price", "Rating"]
```

```python
# EDA
df.describe()
df.info()
```

```python
cleaned_df = df.dropna()
```

```python
# Shifting the cleaned data into a csv file
cleaned_df.to_csv('cleaned_data.csv',index=False)
```
